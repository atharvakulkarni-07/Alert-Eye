<h1>Alert-Eye: Driver Drowsiness Detection System using Python and OpenCV</h1>

<h3>Loading necessary libraries</h3>

```python
import cv2 
from scipy.spatial import distance
import dlib 
import imutils
from imutils import face_utils
from pygame import mixer
```

```python
mixer.init()
mixer.music.load("/Users/atharvakulkarni/myfiles/sleep_detection_system/alarm.wav")
```

```python
def eye_asp_rat(eye):
    A = distance.euclidean(eye[1], eye[5])
    B = distance.euclidean(eye[2], eye[4])
    C = distance.euclidean(eye[0], eye[3])
    ear = (A + B) / (2.0 * C)
    return ear
```

```python
(lstart, lend) = face_utils.FACIAL_LANDMARKS_IDXS["left_eye"]
(rstart, rend) = face_utils.FACIAL_LANDMARKS_IDXS["right_eye"]
```

```python
detect = dlib.get_frontal_face_detector() #builtin function of dlib 
predict = dlib.shape_predictor("/Users/atharvakulkarni/myfiles/sleep_detection_system/shape_predictor_68_face_landmarks.dat") #to predict the 68 landmarks
thresh = 0.3 #to set the threshold
flag = 0
frame_check = 20 #to check the frame
cap = cv2.VideoCapture(0) # to provide the frames 
```

```python
while(True):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    subjects = detect(gray,0)

    for subject in subjects:
        shape = predict(gray, subject)
        shape = face_utils.shape_to_np(shape)

        leftEye = shape[lstart:lend]
        rightEye = shape[rstart:rend]

        leftEAR = eye_asp_rat(leftEye)
        rightEAR = eye_asp_rat(rightEye)

        ear = (leftEAR + rightEAR) / 2.0

        leftEyeHull = cv2.convexHull(leftEye)
        rightEyeHull = cv2.convexHull(rightEye)
        cv2.drawContours(frame, [leftEyeHull], -1, (0, 255, 0), 1)
        cv2.drawContours(frame, [rightEyeHull], -1, (0, 255, 0), 1)

        if ear<thresh:
            flag +=1
            print(flag)
            if(flag > frame_check):
                cv2.putText(frame, "DROWSINESS ALERT!", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                mixer.music.play()
        else: flag =0
        

    if not ret:
        break
    cv2.imshow("Frame", frame)
    
    # Press 'q' on the keyboard to exit the loop and close the window
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

```python
# Release the capture and close the window
cap.release()
cv2.destroyAllWindows()
```
